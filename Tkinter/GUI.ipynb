```python
from tkinter import *
from tkinter import filedialog
from PIL import Image, ImageTk
import cv2
import matplotlib.pyplot as plt
from keras.utils import load_img
from keras.utils import img_to_array
from keras.models import load_model
import numpy as np
import time

####################################################################################################################################

w = Tk()
w.configure(background= '#CDCDCD')
w.title('Image Clasification of MILK')
w.geometry("860x460")
w_icon = Image.open("Icon/MILK.png")
w_icon = ImageTk.PhotoImage(w_icon)
w.iconphoto(False, w_icon)

fontstyle = ("Arial", 12, "bold")
fontstyle_underline = ("Arial", 12, "bold", "underline")
model = load_model('MILK.h5')
dict = {1:'Dutch_lady', 2:'Fami', 3:'Happy_star', 4:'Nuti', 5:'Vinamilk'}

####################################################################################################################################
def open_image():
    file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png;*.jpg;*.jpeg")])
    if file_path:
        image = Image.open(file_path)
        image = image.resize((290,290))
        photo = ImageTk.PhotoImage(image)
        l1.configure(image=photo)
        l1.image = photo
        predict_model(image)
        
def clear_all(): 
    l1.image = None  
    x1 = -10
    x2 = -10
    canvas.coords(canv1, x1, y1, x2, y2)
    canvas.coords(canv2, x1, y1 + do_gian_cach, x2, y2 + do_gian_cach)
    canvas.coords(canv3, x1, y1 + do_gian_cach*2, x2, y2 + do_gian_cach*2)
    canvas.coords(canv4, x1, y1 + do_gian_cach*3, x2, y2 + do_gian_cach*3)
    canvas.coords(canv5, x1, y1 + do_gian_cach*4, x2, y2 + do_gian_cach*4) 
    l4_1.config(text= 'Accuracy: ')
    l5_1.config(text= 'Accuracy: ')
    l6_1.config(text= 'Accuracy: ')
    l7_1.config(text= 'Accuracy: ')
    l8_1.config(text= 'Accuracy: ')    
    capture.release()
    l1.imgtk = None
    l1.configure(image=None)
    
def predict_model(image): 
    img = image.resize((300,300))
    img = img_to_array(img)
    img = img.reshape(1,300,300,3)
    img = img.astype('float32')/255
    prediction = model.predict(img)
    for i in range(1,6):
        value = prediction[0][i]
        #print(dict[i]+':', value)
        canvas_pred(i, value)        
    label = np.argmax(prediction)
    #print(prediction[0][label])
    #print('Result:', dict[label])
    l4_1.config(text= 'Accuracy: ' + str(prediction[0][1]))
    l5_1.config(text= 'Accuracy: ' + str(prediction[0][2]))
    l6_1.config(text= 'Accuracy: ' + str(prediction[0][3]))
    l7_1.config(text= 'Accuracy: ' + str(prediction[0][4]))
    l8_1.config(text= 'Accuracy: ' + str(prediction[0][5]))

def canvas_pred(index, value):
    x1 = 20
    x2 = 20 + value*390
    if index == 1: 
        canvas.coords(canv1, x1, y1, x2, y2)
    if index == 2: 
        canvas.coords(canv2, x1, y1 + do_gian_cach, x2, y2 + do_gian_cach)
    if index == 3: 
        canvas.coords(canv3, x1, y1 + do_gian_cach*2, x2, y2 + do_gian_cach*2)
    if index == 4: 
        canvas.coords(canv4, x1, y1 + do_gian_cach*3, x2, y2 + do_gian_cach*3)
    if index == 5:        
        canvas.coords(canv5, x1, y1 + do_gian_cach*4, x2, y2 + do_gian_cach*4) 
        
def cam_on():
    global capture
    capture = cv2.VideoCapture(0)
    show_frames()
    
def show_frames(): 
    # Get the latest frame and convert into Image
    result, img = capture.read()
    if result:        
        cv2.imwrite("captured_image.jpg", img)
        cv2image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img_cam = Image.fromarray(cv2image)
        # Convert image to PhotoImage
        imgtk = ImageTk.PhotoImage(image = img_cam)
        l1.imgtk = imgtk
        l1.configure(image=imgtk)
        img_pre = Image.open("captured_image.jpg")
        #img_pre = img_pre.transpose(Image.FLIP_LEFT_RIGHT)   
        predict_model(img_pre)
        l1.after(25,show_frames)
        
def cam_off():
    capture.release()
    l1.imgtk = None
    l1.configure(image=None)
    canvas.coords(canv1, x1, y1, x2, y2)
    canvas.coords(canv2, x1, y1 + do_gian_cach, x2, y2 + do_gian_cach)
    canvas.coords(canv3, x1, y1 + do_gian_cach*2, x2, y2 + do_gian_cach*2)
    canvas.coords(canv4, x1, y1 + do_gian_cach*3, x2, y2 + do_gian_cach*3)
    canvas.coords(canv5, x1, y1 + do_gian_cach*4, x2, y2 + do_gian_cach*4) 
    l4_1.config(text= 'Accuracy: ')
    l5_1.config(text= 'Accuracy: ')
    l6_1.config(text= 'Accuracy: ')
    l7_1.config(text= 'Accuracy: ')
    l8_1.config(text= 'Accuracy: ') 
    
auto_mode = False    
def toggle_mode():
    global auto_mode
    # switch to camera on
    if auto_mode==False:
        auto_mode = True     
        cam_on()
    # switch to camera off
    else:
        auto_mode = False        
        cam_off() 

####################################################################################################################################

#Icon
label_1_icon = Image.open("Icon/Label_1.png")
label_1_icon = ImageTk.PhotoImage(label_1_icon) 
button_icon = Image.open("Icon/Button_icon.png")
button_icon = ImageTk.PhotoImage(button_icon)
cam_icon = Image.open("Icon/Camera_icon.png")
cam_icon = ImageTk.PhotoImage(cam_icon)
label_text = Image.open("Icon/Label_text.png")
label_text = ImageTk.PhotoImage(label_text)
import_icon = Image.open("Icon/Import.png")
import_icon = import_icon.resize((30,30))
import_icon = ImageTk.PhotoImage(import_icon)
clear_icon = Image.open("Icon/Clear.png")
clear_icon = clear_icon.resize((25,25))
clear_icon = ImageTk.PhotoImage(clear_icon)

####################################################################################################################################

#Create Frame
frame1 = Frame(w, bg="blue", width=430, height=460)
frame1.place(x = 0, y = 0) 
#Create canvas
canvas = Canvas(w, width=430, height=460, bg="#FFFFFF")
canvas.place(x = 430, y = 0)

label_1 = Label(w, image = label_1_icon, bg = 'blue')
label_1.place(x = 20, y = 60, width=390, height=310)
#Show image or camera 
l1 = Label(w, bg = 'white')
l1.place(x = 40, y = 80, width=350, height=270)

#######################################################################################################################################

#button
bt1 = Button(w, fg = 'blue', image = button_icon, bg = 'blue', command=open_image, bd = 0, activebackground = 'blue')
bt1.place(x = 80, y= 395, height=42, width=157)
bt2 = Button(w, fg = 'blue', image = button_icon, bg = 'blue', command=clear_all, bd = 0, activebackground = 'blue')
bt2.place(x = 255, y= 395, height=42, width=157)
bt3 = Button(w, image = cam_icon, bg = 'blue', command=toggle_mode, bd = 0, activebackground = 'blue')
bt3.place(x = 18, y= 393, height=42, width=42)

#######################################################################################################################################

#Tạo thanh canvas cho từng đối tượng
do_gian_cach = 70       
x1 = -10
y1 = 125
x2 = x1
y2 = y1 + 20         
canv1 = canvas.create_rectangle(x1, y1, x2, y2,outline = '', fill="blue")                                    #Dutch_lady
canv2 = canvas.create_rectangle(x1, y1 + do_gian_cach, x2, y2 + do_gian_cach,outline = '', fill="blue")      #Fami
canv3 = canvas.create_rectangle(x1, y1 + do_gian_cach*2, x2, y2 + do_gian_cach*2,outline = '', fill="blue")  #Happy_star
canv4 = canvas.create_rectangle(x1, y1 + do_gian_cach*3, x2, y2 + do_gian_cach*3,outline = '', fill="blue")  #Nuti
canv5 = canvas.create_rectangle(x1, y1 + do_gian_cach*4, x2, y2 + do_gian_cach*4,outline = '', fill="blue")  #Vinamilk

#######################################################################################################################################

height_text = 35
width_text = 125
l4_2 = Label(w,image = label_text, bg = 'white')
l4_2.place(x = 450, y = 85, height=height_text, width= width_text)
l5_2 = Label(w,image = label_text, bg = 'white')
l5_2.place(x = 450, y = 85 + do_gian_cach, height=height_text, width= width_text)
l6_2 = Label(w,image = label_text, bg = 'white')
l6_2.place(x = 450, y = 85 + 2*do_gian_cach, height=height_text, width= width_text)
l7_2 = Label(w,image = label_text, bg = 'white')
l7_2.place(x = 450, y = 85 + 3*do_gian_cach, height=height_text, width= width_text)
l8_2 = Label(w,image = label_text, bg = 'white')
l8_2.place(x = 450, y = 85 + 4*do_gian_cach, height=height_text, width= width_text)

#Text
x_axis = 465
y_axis = 90
text_bg_color = 'blue'
text_fg_color = 'white'
l4 = Label(w, text = 'Dutch_lady', font = fontstyle, fg = text_fg_color, bg = text_bg_color)
l4.place(x = x_axis, y = y_axis)
l5 = Label(w, text = 'Fami', font = fontstyle, fg = text_fg_color, bg = text_bg_color)
l5.place(x = x_axis, y = y_axis + do_gian_cach)
l6 = Label(w, text = 'Happy_star', font = fontstyle, fg = text_fg_color, bg = text_bg_color)
l6.place(x = x_axis, y = y_axis + 2*do_gian_cach)
l7 = Label(w, text = 'Nuti', font = fontstyle, fg = text_fg_color, bg = text_bg_color)
l7.place(x = x_axis, y = y_axis + 3*do_gian_cach)
l8 = Label(w, text = 'Vinamilk', font = fontstyle, fg = text_fg_color, bg = text_bg_color)
l8.place(x = x_axis, y = y_axis + 4*do_gian_cach)

x_axis_1 = 590
l4_1 = Label(w, text = 'Accuracy: ', font = fontstyle, fg = 'blue', bg = '#FFFFFF')
l4_1.place(x = x_axis_1, y = y_axis)
l5_1 = Label(w, text = 'Accuracy: ', font = fontstyle, fg = 'blue', bg = '#FFFFFF')
l5_1.place(x = x_axis_1, y = y_axis + do_gian_cach)
l6_1 = Label(w, text = 'Accuracy: ', font = fontstyle, fg = 'blue', bg = '#FFFFFF')
l6_1.place(x = x_axis_1, y = y_axis + 2*do_gian_cach)
l7_1 = Label(w, text = 'Accuracy: ', font = fontstyle, fg = 'blue', bg = '#FFFFFF')
l7_1.place(x = x_axis_1, y = y_axis + 3*do_gian_cach)
l8_1 = Label(w, text = 'Accuracy: ', font = fontstyle, fg = 'blue', bg = '#FFFFFF')
l8_1.place(x = x_axis_1, y = y_axis + 4*do_gian_cach)

#text for button
INPUT_text = Label(w, text = 'INPUT', font = fontstyle_underline, fg = 'white', bg = 'blue')
INPUT_text.place(x = 20, y = 20)
RESULT_text = Label(w, text = 'RESULT', font = fontstyle_underline, fg = 'blue', bg = 'white')
RESULT_text.place(x = 450, y = 20)
IMPORT_text = Label(w, image = import_icon, fg = 'blue', bg = 'white')
IMPORT_text.place(x = 145, y = 399)
CLEAR_text = Label(w, image = clear_icon, fg = 'blue', bg = 'white')
CLEAR_text.place(x = 320, y = 401)

w.mainloop()
```
